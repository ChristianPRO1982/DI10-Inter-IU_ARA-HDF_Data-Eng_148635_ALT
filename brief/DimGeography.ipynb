```python
from pathlib import Path

path = Path("../data_adventureWorks/DimGeography2.csv")

with path.open("rb") as f:
    raw = f.read(4096)

print(raw[:200])  # bytes (permet de repérer BOM, séparateur, etc.)

with path.open("r", encoding="utf-8", errors="replace") as f:
    for i in range(1, 21):
        line = f.readline()
        print(f"{i:02d}: {line.rstrip()}")
```

```python
seps = [",", ";", "\t", "|"]
with path.open("r", encoding="utf-8", errors="replace") as f:
    lines = [f.readline() for _ in range(20)]

for s in seps:
    counts = [ln.count(s) for ln in lines if ln]
    print(s, "min/avg/max =", min(counts), sum(counts)/len(counts), max(counts))
```

```python
import pandas as pd

df_head = pd.read_csv(
    path,
    sep=None,            # détection automatique (python engine)
    engine="python",
    on_bad_lines="warn", # affiche les lignes problématiques
    nrows=50,
)
print(df_head.shape)
display(df_head.head())
```

```python
import sweetviz as sv

report = sv.analyze(df_head)
report.show_html("EDA_DimGeography.html")
```
